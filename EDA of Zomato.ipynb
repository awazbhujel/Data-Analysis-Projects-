Merge helps us to combine two data set

```python
final_df = pd.merge(df,df1,on='Country Code',how="left")#on helps on which columns we are combining the data, #how-we use how to specify on left or right table 
```

```python
final_df.head()
```

```python
#to check dtypes
final_df.dtypes
```

```python
final_df.columns
```

Using values and index of Country so that we can display in pie chart

```python
country_names=final_df.Country.value_counts().index
```

```python
country_value = final_df.Country.value_counts().values
```

```python
#pie chart
fig = px.pie(final_df,country_value,labels=country_names,hover_name=country_names,
             title="Country that uses Zomato")
fig.show()
```

```python
plt.pie(country_value[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Observation:from this piechart we observe that zomato maximum records or transaction are from india after that usa and united Kingdom.

```python
final_df.columns
```

```python
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

Observation:

1.when Ratings is between 4.5 to 4.9---> excellent

2.when Ratings is between 4.0 to 4.4 --> very good

3.when Ratings is between 3.5 to 3.9 --> good

4.when Ratings is between 2.5 to 3.4 --> average

5.when Ratings is between 1.8 to 2.4 --> poor

6.5.when Ratings is 0--> Not Rated


```python
ratings.head()
```

```python
#barplot
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)
sns.barplot(data=ratings,x="Aggregate rating",y="Rating Count")
```

```python
sns.barplot(data=ratings,x="Aggregate rating",y="Rating Count",palette=["blue","red","orange","yellow","green","green"],hue="Rating color")
```

Observations:

1.Not rated count is very high

2.Maximum no.of ratting are between 2.6 to 3.6

```python
##count plot
sns.countplot(data=ratings,x="Rating color",palette=["blue","red","orange","yellow","green","green"])
```

```python
final_df.columns
```

```python
#find the countries name who has gives 0 rating
final_df[final_df["Aggregate rating"]==0.0].groupby("Country").size().reset_index()
```

```python
#another method
final_df.groupby(["Aggregate rating","Country"]).size().reset_index().head(5)
```

Observation:

Maximum number of 0 ratings are from Indian Customers

```python
#find out which currency is used by which country
final_df[["Currency","Country"]].groupby(["Currency","Country"]).size().reset_index()
```

```python
#which country has online delivary options
final_df[final_df["Has Online delivery"]=="Yes"].groupby("Country").size().reset_index()
```

Observations:

1.Online Delivery are available in India and Uae

*Create a pie chart for top 5 city distribution 

```python
city_values=final_df.City.value_counts().values
city_index=final_df.City.value_counts().index
```

```python
plt.pie(city_values[:5],labels=city_index[:5],autopct="%.2f%%")
plt.show()
```

```python
##find the top 10 cuisines
cuisines_values=final_df['Cuisines'].value_counts().sort_values(ascending=False).head(10).values
cuisines_index=final_df['Cuisines'].value_counts().sort_values(ascending=False).head(10).index
```

```python
plt.pie(cuisines_values,labels=cuisines_index)
plt.show()
```

Concusions:

North Indian and chinese food are most famous dishes in Zomato
